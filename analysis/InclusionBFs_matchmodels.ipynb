```r
rawRTData <- read.csv('SC_ANOVA_RT.csv')
rawRTData <- within(rawRTData, {subject <- factor(subject)
    BlockType <- factor(BlockType)
    Congruency <- factor(Congruency)
    BlockOrder <- factor(BlockOrder)
    Feedback <- factor(Feedback)
})
SC_RT_runANOVA <- anovaBF(RT ~ BlockType*Congruency*BlockOrder*Feedback + subject, data=rawRTData, whichRandom="subject")
print(bayesfactor_inclusion(SC_RT_runANOVA, match_models=TRUE))

#

rawRTData_T <- read.csv('SCT_ANOVA_RT.csv')
rawRTData_T <- within(rawRTData_T, {subject <- factor(subject)
    BlockType <- factor(BlockType)
    Congruency <- factor(Congruency)
    BlockOrder <- factor(BlockOrder)
    Feedback <- factor(Feedback)
})
SCT_RT_runANOVA <- anovaBF(RT ~ BlockType*Congruency*BlockOrder*Feedback + subject, data=rawRTData_T, whichRandom="subject")
print(bayesfactor_inclusion(SCT_RT_runANOVA, match_models=TRUE))

#

rawACCData <- read.csv('SC_ANOVA_ACC.csv')
rawACCData <- within(rawACCData, {subject <- factor(subject)
    BlockType <- factor(BlockType)
    Congruency <- factor(Congruency)
    BlockOrder <- factor(BlockOrder)
    Feedback <- factor(Feedback)
})
SC_ACC_runANOVA <- anovaBF(Accuracy ~ BlockType*Congruency*BlockOrder*Feedback + subject, data=rawACCData, whichRandom="subject")
print(bayesfactor_inclusion(SC_ACC_runANOVA, match_models=TRUE))

#

rawACCData_T <- read.csv('SCT_ANOVA_ACC.csv')
rawACCData_T <- within(rawACCData_T, {subject <- factor(subject)
    BlockType <- factor(BlockType)
    Congruency <- factor(Congruency)
    BlockOrder <- factor(BlockOrder)
    Feedback <- factor(Feedback)
})
SCT_ACC_runANOVA <- anovaBF(Accuracy ~ BlockType*Congruency*BlockOrder*Feedback + subject, data=rawACCData_T, whichRandom="subject")
print(bayesfactor_inclusion(SCT_ACC_runANOVA, match_models=TRUE))
```
